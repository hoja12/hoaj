```python
def check_contract(val):
    if 'On Loan' in val:
        return'On Loan'
    elif 'Free' in val:
        return 'Free Agent'
    elif '~' in val:
        return 'Under Contract'
    else:
        return 'Unknown'
```

```python
data['contract_type'] = data['Contract'].apply(lambda x : check_contract(x))
```

```python
data[['contract_type']]
```

```python
data.contract_type.value_counts()
```

```python
#column for contract start
data['Contract_start'] = data['Joined'].apply(lambda x: x[-5:])
```

```python
#to create column for contract end
def value(val):
    if 'Loan' in val:
        val = val[7:12]
        val = val.strip(' ') # I addedn this line because some whitespace was present originally
        return val
    else:
        val = val[-4:]
        return val
        
data['Contract_end'] = data['Contract'].apply(lambda x : value(x))
```

```python
data[['Contract_start','Contract_end']]
```

```python
#drop column contract
data.drop('Contract', axis = 1, inplace = True)
```

```python

data['Club'] = data['Club'].apply(lambda x: x[4:])
data[['Club']].head()
```

```python
#check for inches values in height
inches = data[data.Height.str.contains("'")]
inches_height = inches['Height']
inches_height.sample(5)
```

```python
def inches_to_cm(val):
    if "'" in val:
        val = val.strip('"')
        feet,inches = val.split("'") 
        inches_val = round((float(feet)*30.48 + float(inches)* 2.54),2)
        return inches_val
    else:
        val = val.strip('cm') 
        return val
    
data['Height_cm'] = data['Height'].apply(lambda x: inches_to_cm(x))
```

```python
data.Height_cm = data.Height_cm.astype(float)
```

```python
data.drop('Height', axis = 1, inplace = True)
```

```python
data[['Height_cm']].head()
```

```python
def lbs_to_kg (val):
    if 'lbs' in val:
        val = val.strip('lbs')
        kg_val = round((float(val)* 0.453592),2)
        return kg_val
    else:
        val = val.strip('kg')
        return val
    
data['weight_kg'] = data['Weight'].apply(lambda x : lbs_to_kg(x))
```

```python
data['weight_kg'] = data['weight_kg'].astype(float)
```

```python
data.drop('Weight', axis = 1, inplace = True)
```

```python
data[['weight_kg']].head()
```

```python
from datetime import datetime

def date_time (val):
    change_date = datetime.strptime(val, '%b %d, %Y')
    return change_date
    
data['Joined'] = data['Joined'].apply(lambda x : date_time (x))
```

```python
data[['Joined']].head()
```

```python
data['Value'] = data['Value'].str.strip('€') #this removes the € symbol
data['Release Clause'] = data['Release Clause'].str.strip('€')

def change_value(val):
    if 'M' in val:
        val = val.strip('M') #this removes the M symbol
        million = float(val) * 1000000
        return million
    else:
        val = val.strip('K')
        thousand = float(val) * 1000
        return thousand
    
data['Value_euro'] = data['Value'].apply(lambda x: change_value(x)) 
data['Release Clause_euro'] = data['Release Clause'].apply(lambda x: change_value(x))
```

```python
data.drop(['Value', 'Release Clause'], axis = 1, inplace = True)
```

```python
data[['Value_euro', 'Release Clause_euro']].head()
```

```python
data['Wage'] = data['Wage'].str.strip('€')

def change_value(val):
    if 'K' in val:
        val = val.strip('K')
        thousand = float(val) * 1000
        return thousand
    else:
        val = val
        return val
    
data['Wage_euro'] = data['Wage'].apply(lambda x: change_value(x))
```

```python
data['Wage_euro'] = data['Wage_euro'].astype(float)
```

```python
data.drop('Wage', axis = 1, inplace = True)
```

```python
data[['Wage_euro']].head()
```

```python
#removing star in columns SM, W/F and IR
data[['W/F','SM','IR']] = data[['W/F','SM','IR']].apply(lambda x: x.str.strip('★'))

# this changes the datatype of these columns
data[['W/F','SM','IR']] = data[['W/F','SM','IR']].astype(int)
```

```python
data[['SM', 'W/F','IR']].head()
```

```python
def goalkeeping(val):
    val = (val/500) * 100
    return val

data['GoalKeeping_percentage'] = data['Goalkeeping'].apply(lambda x: goalkeeping(x))
```

```python
data.drop(['Attacking','Skill','Movement', 'Power', 'Mentality', 'Defending', 'Goalkeeping' ], axis = 1, inplace = True )
```

```python
data = data.rename(columns = {'↓OVA': 'OVA', 'W/F': 'WF', 'A/W': 'AW', 'D/W': 'DW'})

# Change the columns to lowercase
data.columns = data.columns.str.lower()

# Replace whitespace with underscore
data.columns = data.columns.str.replace(' ','_')
```
